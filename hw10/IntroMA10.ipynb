## Найти интегралы:


```python
# Задача 1

# Задаём символьную переменную
x, c = sym.symbols('x C')

# Задаём функцию
func = (x**4 + x**2 - 6*x) / x**3

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 2
# Задаём функцию
func = sym.cos(2*x)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 3
# Задаём функцию
func = 1/(3*x+2)**4
# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,sym.factor(a.doit())+c)
```

```python
# Задача 4

# Задаём функцию
func = (5*x - 1)/ sym.sqrt(4-x**2)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 5

# Задаём функцию
func = 1/ (sym.sqrt(x)*(1+sym.sqrt(x)))

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 6

# Задаём функцию
func = x**2*sym.cos(x)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 7

# Задаём функцию
func = sym.atan(x)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,a.doit()+c)
```

```python
# Задача 8

# Задаём функцию
func = 1/(x-1)**5

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,sym.factor(a.doit())+c)
```

```python
# Задача 9

# Задаём функцию
func = (x+6)/(x**2-2*x+17)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,sym.factor(a.doit())+c)
```

```python
# Задача 10

# Задаём функцию
func = x/((x**2-1)*(x**2+1))

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a,sym.factor(a.doit())+c)
```

```python
# Задача 11

# Задаём функцию
func = sym.cbrt(x)/(sym.cbrt(x**2)-sym.sqrt(x))

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a, a.doit()+c)
```

```python
# Задача 12

# Задаём функцию
func = sym.cbrt(x)*sym.cbrt(1+3*sym.cbrt(x**2))

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a, a.doit()+c)
```

```python
# Задача 13

# Задаём функцию
func = 1/(sym.sin(x)*sym.sin(2*x))

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a, sym.factor(a.doit())+c)
```

```python
# Задача 14

# Задаём функцию
func = sym.cos(5*x) * sym.cos(3*x)

# Вычисляем
a=sym.Integral(func,x)
sym.Eq(a, sym.trigsimp(a.doit())+c)
```

```python
# Задача 15

# Задаём функцию
func = 2*x + sym.sin(2*x)

# Вычисляем
a=sym.Integral(func, (x, 0, sym.pi))
sym.Eq(a, sym.trigsimp(a.doit()))
```

```python
# Задача 16

# Задаём функцию
func = sym.sqrt(4*x -2)

# Вычисляем
a=sym.Integral(func, (x, 1/2, 1))
sym.Eq(a, a.doit())
```

```python
# Задача 17

# Задаём функцию
func = sym.E**(-4*x)

# Вычисляем
a=sym.Integral(func, (x, 0, sym.oo))
sym.Eq(a, a.doit())
```
