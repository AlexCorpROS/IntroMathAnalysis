## 1. Найти
 $\frac{dz}{dt}$, если $z=z(x;y), x=x(t), y=y(t)$:  
### $z=x^2+y^2+xy, \quad x=a\,sin\,t,\, y=a\,cos\,t$  


```python
# Задаём символьную переменную
x, y, t, a = sym.symbols('x y t a')
# Задаём функцию z
func_z = x**2 + y**2 + x*y
# Задаём функцию x
func_x = a * sym.sin(t)
# Задаём функцию y
func_y = a * sym.cos(t)

# Вычисляем частные производные по x, y
dz_dx = sym.diff(func_z, x)
dz_dy = sym.diff(func_z, y)
# Вычисляем полные производные по t
dx_dt = sym.diff(func_x, t)
dy_dt = sym.diff(func_y, t)

```

```python
print(f'{(dz_dx*dx_dt + dz_dy*dy_dt)} ')
```

Ответ.

$\dfrac{dz}{dt}=\,$ $-a*(x + 2*y)*sin(t) + a*(2*x + y)*cos(t) $

## 2. Для данных
 $z=f(x;y), x=x(u;v), y=y(u;v)$ найти $\frac{\partial z}{\partial u}, \frac{\partial z}{\partial v}$ и $dz$:  
### $z=x^3+y^3, \quad x=uv,\,y=\frac{u}{v}$



```python
# Задаём символьную переменную
u, v, du, dv = sym.symbols('u v du dv')

# Задаём функцию z
func_z = x**3 + y**3

# Задаём функцию x
func_x = u * v

# Задаём функцию y
func_y = u / v

# Подставим значения x и y в формулу
z = func_z.subs({x: func_x, y: func_y})

# Вычисляем частные производные
z_v = sym.diff(z, v)
z_u = sym.diff(z, u)

# Вычисляем dz
dz = z_u*du + z_v*dv

# Выводим результат
print(f'{(z_v)}')
print(f'{(z_u)}')
print(f'{(dz)}')
```

Ответ.

$\dfrac{\partial z}{\partial v}=\;$ $3*u^3*v^2 - \dfrac{3*u^3}{v^4} $

$\dfrac{\partial z}{\partial u}=\;$ $ 3*u^2*v^3 + \dfrac{3*u^2}{v^3} $

$dz=\;$ $ du*(3*u^2*v^3 + \dfrac{3*u^2}{v^3}) + dv*(3*u^3*v^2 - \dfrac{3*u^3}{v^4})  $

## 3. Найти производные $y'(x)$ неявных функций, заданных уравнением:
  
 $xe^{2y}-y\,ln\,x=8$



```python
# Задаём символьную переменную
x, y = sym.symbols('x y')

# Задаём функцию z
func = x*sym.E**(2*y) - y*sym.ln(x) - 8

# Вычисляем производную неявной функции
result = -(sym.diff(func, x)/sym.diff(func, y))

# Выводим результат

print(f'{(sym.simplify(result))}')
```

Ответ.

$y'(x)=\;$ $ \dfrac{-xe^{2y} + y} {x*(2xe^{2y} - log(x))} $

## 4. Составить уравнение касательной прямой и нормали к кривой
 $y=y(x)$, заданной уравнением $F(x; y)=0$ в точке $M_0(x_0;y_0)$:  

 $x^3y-y^3x=6, \quad M_0(2;1)$



```python
# Задаём символьную переменную
x, y = sym.symbols('x y')

# Задаём функцию z
func = x**3*y + y**3*x - 6

# Точка (x, y)
M = (2, 1)

# Вычисляем производную неявной функции
k_def = -(sym.diff(func, x)/sym.diff(func, y))

# Вычисляем коэффициент k в точке M(x, y)
k_ = k_def.subs({x:M[0], y:M[1]})


# Формула касательной y - y0 = k * (x -x0)
print(f'{(k_)}(x-{M[0]})')

# Формула нормали y - y0 = 1/k * (x -x0)
print(f'{(1/k_)}(x-{M[0]})')
```

Уравнение касательной $ y-1 = -\dfrac{13}{14}(x-2) $

Формула нормали $ y - 1 = -\dfrac{14}{13}(x-2) $

## 5. Для данных функций найти требуемую частную производную или дифференциал:

1. $z=sin\,x \,\cdot\, sin\,y, \quad d^2z$  


```python
# Задаём символьную переменную
x, y, dx, dy = sym.symbols('x y dx dy')

# Задаём функцию z
z = sym.sin(x) * sym.sin(y)

# Вычисляем частные производные второго порядка
z2_x = sym.diff(z, x, 2)
z2_y = sym.diff(z, y, 2)

# Вычисляем смешанную производную
z_x = sym.diff(z, x)
z_xy = sym.diff(z_x, y)

# Вычисляем дифференциал второго порядка
d2_z = z2_x*dx**2 + 2*z_xy*dx*dy + z2_y*dy**2

# Выводим результат
print(f'{(d2_z)}')
```

$d^2z = -dx^2*sin(x)*sin(y) + 2*dx*dy*cos(x)*cos(y) - dy^2*sin(x)*sin(y) $  

2.  $z=xy+sin(x+y), \quad \frac{\partial ^2z}{\partial x^2}$  


```python
# Задаём символьную переменную
x, y = sym.symbols('x y')

# Задаём функцию z
z = x*y + sym.sin(x+y)

# Вычисляем производную второго порядка по x
result = sym.Derivative(z, x, 2)

# Выводим результат
sym.Eq(result, result.doit())
```
