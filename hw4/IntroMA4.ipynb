### Тема “Предел функции”

### 1. Задание
Предложить пример функции, не имеющей предела в нуле и в бесконечностях.
$$f(x)=\sin(x)+\sin(\frac{1}{x})$$

$f_1(x)=\sin{x}$ - периодическая функция, поэтому у нее не существует предела на бесконечностях. <br><br>
$f_2(x)=\sin{\frac{1}{x}}$ - рассмотрим последовательность $x_n=\frac{2}{\pi(2n+1)}$, тогда<br>
$f_2(x_n)=\sin{\frac{\pi(2n+1)}{2}}=\sin{(\pi n+\frac{\pi}{2})}=\cos{(\pi n)}=(-1)^n$, получаем <br>
$lim_{n\to \infty}x_n=lim_{n\to \infty}\frac{2}{\pi(2n+1)}=0$, т.е. $x_n\to0$, а $f_2(x_n)$ не имеет предела при $n\to \infty$.<br>
Значит и исследуемая функция $f_2(x)=\sin{\frac{1}{x}}$ не имеет предела при $x\to 0$.<br><br>
Линейная комбинация этих двух функций и дает нам пример функции, не имеющей предела в нуле и бесконечностях.



```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sym
import warnings
import math

warnings.filterwarnings('ignore')
%matplotlib inline

x_l = sym.symbols('x')
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1001)
y = np.sin(x)+np.sin(1/x)
plt.figure(figsize=(10,5))
plt.plot(x, y, label='sin(x)+sin(1/x)', c='g')
plt.plot([-2*np.pi, 2*np.pi], [0, 0], color='r', linestyle='--', alpha=0.5)
plt.plot([0, 0], [-2, 2], color='r', linestyle='--', alpha=0.5)
plt.axis([-2*np.pi, 2*np.pi, -2, 2])
plt.legend()
plt.grid()
plt.show()
```

### 4. Задание

Найти предел:

$$a.\lim_{x\to0}\frac{3x^3-2x^2}{4x^2}$$

Аналитическое решение:

$\lim_{x\to0}\frac{3x^3-2x^2}{4x^2}=\lim_{x\to0}\frac{x^2(3x-2)}{4x^2}=\lim_{x\to0}\frac{3x-2}{4}=\frac{-2}{4}=-0.5$

```python
# Численное решение
func_a = (3*x_l**3 - 2*x_l**2)/(4*x_l**2)
print(sym.limit(func_a, x_l, 0))
```

$$b^*.\lim_{x\to0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$$

### Аналитическое решение:

$\lim_{x\to0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}=\lim_{x\to0}\frac{(\sqrt{1+x}-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{(\sqrt[3]{1+x}-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}=\lim_{x\to0}\frac{(\sqrt{1+x}-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{(1+x-1)}=\lim_{x\to0}\frac{(\sqrt{1+x}-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)(\sqrt{1+x}+1)}{x(\sqrt{1+x}+1)}=\lim_{x\to0}\frac{(1+x-1)(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{x(\sqrt{1+x}+1)}=\lim_{x\to0}\frac{x(\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1)}{x(\sqrt{1+x}+1)}=\lim_{x\to0}\frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{\sqrt{1+x}+1}=\frac{3}{2}=1.5$

```python
# Численное решение
func_b = (sym.sqrt(1+x_l)-1)/((1+x_l)**(1/3)-1)
print(sym.limit(func_b, x_l, 0))
```

$$c^*.\lim_{x\to\infty}(\frac{x+3}{x})^{4x+1}$$

### Аналитическое решение:

$\lim_{x\to\infty}(\frac{x+3}{x})^{4x+1}=\lim_{x\to\infty}(1+\frac{3}{x})^{4x+1}=\lim_{x\to\infty}((1+\frac{3}{x})(1+\frac{3}{x})^{4x})=\lim_{x\to\infty}(1+\frac{3}{x})*\lim_{x\to\infty}(1+\frac{3}{x})^{4x}=1*\lim_{x\to\infty}(1+\frac{3}{x})^{4x}=|_{y=\frac{x}{3}}|=\lim_{y\to\infty}(1+\frac{1}{y})^{12y}=\lim_{y\to\infty}((1+\frac{1}{y})^{y})^{12}=|_{2\underline{}{замечательный}\underline{}{предел}}|=e^{12}$

```python
# Численное решение
func_с = ((x_l+3)/(x_l))**(4*x_l+1)
print(sym.limit(func_с, x_l, sym.oo))
```

### Тема “Теоремы о пределах”

### 1. Задание
Найти предел:

$$a.\lim_{x\to0}\frac{\sin{(2x)}}{4x}$$

Аналитическое решение:

$\lim_{x\to0}\frac{\sin{(2x)}}{4x}=\frac{1}{2}\lim_{x\to0}\frac{\sin{2x}}{2x}=|_{y=2x}|=\frac{1}{2}\lim_{y\to0}\frac{\sin{y}}{y}=|_{1\underline{}{замечательный}\underline{}{предел}}|=\frac{1}{2}=0.5$

```python
# Численное решение
func_a = (sym.sin(2*x_l))/(4*x_l)
print(sym.limit(func_a, x_l, 0))
```

$$b.\lim_{x\to0}\frac{x}{\sin{(x)}}$$

Аналитическое решение:

$\lim_{x\to0}\frac{x}{\sin{(x)}}=\lim_{x\to0}(\frac{\sin{(x)}}{x})^{-1}=(\lim_{x\to0}\frac{\sin{(x)}}{x})^{-1}=|_{1\underline{}{замечательный}\underline{}{предел}}|=(1)^{-1}=1$

```python
# Численное решение
func_b = (x_l)/(sym.sin(x_l))
print(sym.limit(func_b, x_l, 0))
```

$$c.\lim_{x\to0}\frac{x}{\arcsin{(x)}}$$

Аналитическое решение:

$\lim_{x\to0}\frac{x}{\arcsin{(x)}}=|_{x=\sin{y}}|=\lim_{y\to0}\frac{\sin{(y)}}{y}=|_{1\underline{}{замечательный}\underline{}{предел}}|=1$

```python
# Численное решение
func_с = (x_l)/(sym.asin(x_l))
print(sym.limit(func_с, x_l, 0))
```

$$d.\lim_{x\to\infty}(\frac{4x+3}{4x-3})^{6x}$$

Аналитическое решение:

$\lim_{x\to\infty}(\frac{4x+3}{4x-3})^{6x}=\lim_{x\to\infty}(\frac{4x-3+6}{4x-3})^{6x}=\lim_{x\to\infty}(1+\frac{6}{4x-3})^{6x}=|^{y=\frac{4x-3}{6}}_{x=\frac{6y+3}{4}}|=\lim_{y\to\infty}(1+\frac{1}{y})^{6*\frac{6y+3}{4}}=\lim_{y\to\infty}(1+\frac{1}{y})^{9y+\frac{9}{2}}=\lim_{y\to\infty}(1+\frac{1}{y})^{\frac{9}{2}}*(\lim_{y\to\infty}(1+\frac{1}{y})^y)^9=|_{2\underline{}{замечательный}\underline{}{предел}}|=1*(e)^9=e^9$

```python
# Численное решение
func_d = ((4*x_l+3)/(4*x_l-3))**(6*x_l)
print(sym.limit(func_d, x_l, sym.oo))
```

$$e^*.\lim_{x\to\infty}\frac{\sin{x}+\ln{x}}{x}$$

Аналитическое решение:

$\lim_{x\to\infty}\frac{\sin{x}+\ln{x}}{x}=\lim_{x\to\infty}\frac{\sin{x}}{x}+\lim_{x\to\infty}\frac{\ln{x}}{x}$<br><br>
$\sin(x)$-ограниченная функция ($\sin{x}\in[-1,1] \forall x\in\mathbb{R}$)$\Rightarrow$ при $x\to\infty$: $\frac{\sin{x}}{x}\to0$ $\Rightarrow$ $\lim_{x\to\infty}\frac{\sin{x}}{x}=0$<br><br>
Функция $f(x)=x$ растет сильно быстрее, чем функция $f(x)=\ln{x}$ (это видно на графике снизу; также это можно доказать, взяв производные, но мы их еще не проходили в этом курсе), поэтому при $x\to\infty$: $\frac{\ln{x}}{x}\to0$ $\Rightarrow$ $\lim_{x\to\infty}\frac{\ln{x}}{x}=0$<br>
<br>
$\lim_{x\to\infty}\frac{\sin{x}+\ln{x}}{x}=0+0=0$

```python
x = np.linspace(1, 1000, 101)
y = x
y2 = np.log(x)
plt.plot(x, y, label='x', c='g')
plt.plot(x, y2, label='ln(x)', c='orange')
plt.axis([1, 1000, 1, 1000])
plt.legend()
plt.show()
```

```python
# Численное решение
func_e = (sym.sin(x_l)+sym.ln(x_l))/(x_l)
print(sym.limit(func_e, x_l, sym.oo))
```

$$f^*.\lim_{x\to0}\frac{\sin{(x)}+\ln{(x)}}{x}$$

Аналитическое решение:


$\lim_{x\to0}\frac{\sin{x}+\ln{x}}{x}=\lim_{x\to0}\frac{\sin{x}}{x}+\lim_{x\to0}\frac{\ln{x}}{x}$<br><br>
$\lim_{x\to0}\frac{\sin{x}}{x}=|_{1\underline{}{замечательный}\underline{}{предел}}|=1$<br><br>
$\lim_{x\to0}\frac{\ln{x}}{x}$ - посмотрим на график функции $f(x)=\frac{\ln{x}}{x}$(внизу)<br>
Функция $f(x)=\ln{x}$ не определена в 0, при этом она убывает при $x\to0_{+}$ (то есть на интервале $(0,+\infty)$ сама функция возрастает, но если смотреть в другом направлении, то как бы убывает, уменьшается при уменьшении x), т.е. при $x\to0$ $f(x)=\ln{x}\to-\infty$, а мы пытаемся поделить ее еще на 0, т.е. при $x\to0$ функция $f(x)=\frac{\ln{x}}{x}$ очень быстро "убегает" на $-\infty$ $\Rightarrow$ $\lim_{x\to0}\frac{\ln{x}}{x}=-\infty$<br><br>
$\lim_{x\to0}\frac{\sin{x}+\ln{x}}{x}=1+(-\infty)=-\infty$

```python
x = np.linspace(10**(-7), 0.5, 1001)
y = np.log(x) / x
y2 = np.log(x)
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
fig.set_size_inches(6, 10)
fig.subplots_adjust(hspace=0.25, wspace=0.3)
ax1.plot(x, y, label='ln(x)/x', c='g')
ax1.set_xlim([0, 0.5])
ax1.set_ylim([-1000, 1])
ax2.plot(x, y, label='ln(x)/x', c='g')
ax2.plot(x, y2, label='ln(x)', c='orange')
ax2.set_xlim([0, 0.5])
ax2.set_ylim([-100, 1])
plt.legend()
plt.show()
```
